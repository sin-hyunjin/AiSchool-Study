```python
# 포켓몬 사이트 열기
driver = wb.Chrome()
driver.get('https://www.pokemonkorea.co.kr/pokedex#pokedex_1')
```

```python
# 속성값 추출을 위해서 bs 객체화 
# bs(페이지정보, 파싱방법)
soup = bs(driver.page_source,'lxml')
soup
```

```python
# 첫번째 포켓몬(이상해씨) 태그추출 
img = soup.select('img.img-fluid')
img.text
# .text 요소에서 콘텐츠를 추출할때 사용
# img태그는 홀태그로써 콘텐츠를 가지지 않는다!
# 요소에서 속성을 추출!! 
```

- 수집한 요소에서 콘텐츠가 아닌 속성을 추출하는 방법!

```python
# 요소['속성명']
img[0]['src']
# 출력된정보 : 이미지의 경로 주소 
```

```python
# 파일의 경로를 활용하여 이미지 파일로 저장 
# urlretrieve(이미지의 경로, '저장할위치/파일명.확장자')
urlretrieve(img[0]['src'],'/Users/sinhyeonjin/Downloads/포켓몬도감/이상해씨.jpg')
```

```python
# 순수한 src 주소를 담은 리스트
img_list = []
for i in img:
    img_list.append(i['src'])
```
