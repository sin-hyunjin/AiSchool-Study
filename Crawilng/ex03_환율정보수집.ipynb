```python
# 라이브러리 불러오기
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 1. req를 활용해서 네이버 환율사이트 오청
# * 응답코드를 확인하고, 변수에 담자.
url = 'https://finance.naver.com//marketindex/exchangeList.naver'
res = req.get(url)
```

```python
# 2. bs을 통해서 html형태의 데이터로 변형
soup = bs(res.text, 'lxml')
soup
```

```python
# 3. 환율정보 중 국가명을 수집
soup.select('td.tit > a')
```

## iframe 태그
- 화면상에 보여만 주고, 실제 원본데이터는 다른사이트에 존재 하는 경우
- 데이터 수집이 불가능 => 원본데이터가 있는 주소로 이동해서, 수집

```python
# iframe태그의 원본데이터는 src값에 존재한다.
# 경로가 상대경로인 경우에는 반드시 절대경로로 변경 후, 재요청
# /marketindex/exchangeList.naver = 상대경로
# https://finance.naver.com//marketindex/exchangeList.naver = 절대경로
# * req로 요청을 보냈을때, 응답코드가 200번이다. 그러나 올바른 데이터가 출력이 되지 않는 경우는 반드시 iframe여부를 검사해라
```
