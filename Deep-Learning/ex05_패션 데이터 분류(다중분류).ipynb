### 회귀
- loss = MSE (평균제곱오차)
- 출력층 (가장 마지막 레이어) - 뉴런 수 : 1
- 활성화 함수 : linear(항등함수 y=x)

### 분류
- 이진분류
  - loss = binary_crossentropy
  - 출력층 (가장 마지막 레이어) - 뉴런 수 : 1
  - 활성화 함수 : sigmoid(S곡선 형태, 0~1)

- 다중분류
  - loss = categorical_crossentropy
  - 출력층(가장 마지막 레이어) - 뉴런 수 : 클래스 개수
  - 활성화 함수 : softmax( 0~1, 총합 1)


- 다중분류
  - 정답 데이터 : one-hot (확률 정보)
    - loss : categorical_crossentropy
  - 정답 데이터 : 자체 (원-핫X)
    - loss : spare_categorical_crossentropy
  둘 다 성능의 차이 xXXX 그냥 방법 1, 2 벗 맞춰서 써줘야 함

```python
from tensorflow.keras.datasets import fashion_mnist
```

```python
(X_train, y_train),( X_test, y_test) = fashion_mnist.load_data()
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Flatten
from tensorflow.keras.utils import to_categorical
```

```python
# 패션데이터 출력(imshow)
plt.imshow(X_train[0])
```

```python
# 라벨 종류 파악
np.unique(y_train),
```

```python
y_train_one_hot = to_categorical(y_train)
y_test_one_hot = to_categorical(y_test)
```

```python
# 1. 모델 구조 설계
fm_model = Sequential()
fm_model.add (InputLayer(input_shape=(28,28))) # 입력층
fm_model.add(Flatten()) # 입력된 데이터를 1차원으로 평평하게 만들어주는 층
fm_model.add(Dense(units = 400, activation='sigmoid')) # 중간층(은닉층)
fm_model.add(Dense(units = 234, activation='sigmoid'))
fm_model.add(Dense(units = 125, activation='sigmoid'))
fm_model.add(Dense(units = 32, activation='sigmoid'))
fm_model.add(Dense(units = 10, activation='softmax'))# 출력층
# 다중분류 출력층은 무조건 units = 10 / activation(활성화함수) = 'softmax'값을 가져야 함!!
```

```python
# 2. 모델 학습/평가 방법 설정
fm_model.compile (loss = 'categorical_crossentropy',
                     optimizer = 'adam',
                     metrics = ['accuracy'])
```

```python
fm_history = fm_model.fit(X_train, y_train_one_hot, validation_split=0.2, epochs=100)
```

```python
plt.figure(figsize=(15,5))
plt.plot(fm_history.history['accuracy'], label= 'acc')
plt.plot(fm_history.history['val_accuracy'], label= 'val_acc')
plt.legend()
# 바탕부분이 검정색이라 검정색이 젤 많은
plt.show()
```

```python
# 4, 모델 평가
fm_model.evaluate(X_test, y_test_one_hot)
```

```python
# 모델 저장
fm_model.save('fashion_model.h5')
```

### 베스트 모델 찾기

```python
# 1. 모델 구조 설계
fm_model = Sequential()
fm_model.add (InputLayer(input_shape=(28,28))) # 입력층
fm_model.add(Flatten()) # 입력된 데이터를 1차원으로 평평하게 만들어주는 층
fm_model.add(Dense(units = 400, activation='sigmoid')) # 중간층(은닉층)
fm_model.add(Dense(units = 234, activation='sigmoid'))
fm_model.add(Dense(units = 125, activation='sigmoid'))
fm_model.add(Dense(units = 32, activation='sigmoid'))
fm_model.add(Dense(units = 10, activation='softmax'))# 출력층
# 다중분류 출력층은 무조건 units = 10 / activation(활성화함수) = 'softmax'값을 가져야 함!!
```

```python
# 2. 모델 학습/평가 방법 설정
fm_model.compile (loss = 'categorical_crossentropy',
                     optimizer = 'adam',
                     metrics = ['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
import os
```

```python
# 모델만 저장시킬 폴더 생성
!pwd
MODEL_DIR = './model'

# 만약에 폴더가 없다면 생성
if not os.path.exists(MODEL_DIR):
  os.mkdir(MODEL_DIR)
```

```python
# 에포크 + 성능을 한 번에 확인할 수 있는, 성능이 개선된 모델들을 저장
# 정규표현식 : 저장될 모델명의 정규식 설정
model_path = MODEL_DIR + "{epoch:02d}-{val_loss:.4f}.hdf5"

# monitor : 모니터링할 기준 값(저장 기준) (val_acc, loss 등)
# verbose : 진행 내용 출력 여부 (1:출력, 0:출력X)
# save_best_only=True : 이전보다 성능이 개선 되었을 경우에만 저장
checkpointer = ModelCheckpoint(filepath=model_path, monitor = 'val_loss',
                verbose=1, save_best_only=True)
```

```python
# 3. 모델 학습 및 학습결과 시각화
fm_history = fm_model.fit(X_train, y_train_one_hot, validation_split=0.2, epochs=100, callbacks = [checkpointer])
```

```python
import numpy as np
```

```python
y_loss = fm_history.history['loss']
y_vloss = fm_history.history['val_loss']
x_len = np.arange(len(y_vloss))

plt.plot(x_len, y_loss, "o", c="red", markersize=3)
plt.plot(x_len, y_vloss, "o", c="blue", markersize=3)
plt.show()
```

```python
# 저장된 모델 불러오기
modelpath = MODEL_DIR + '98-0.5096.hdf5'

save_model= load_model(modelpath)
```

```python
# one hot으로 학습했음 one hot으로 테스트
save_model.evaluate(X_test,y_test_one_hot)
```
