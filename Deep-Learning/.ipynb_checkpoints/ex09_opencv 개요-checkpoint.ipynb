### Gray(흑백이미지), Binary(이진이미지)
- gray(흑백이미지): 0~255 픽셀로 이루어진 이미지 0에 가까울수록 검정, 255에 가까울수록 흰색
- binary(이진이미지): 0 또는 1로 구성된 이미지
    - 얼굴만 따로 떼어내는 등 필터링이 필요한 이미지 > 필요한 부분만 떼어올때 사용한다~ 

### gray 이미지 만들기
- 두가지 방법
    - 1. imread() 를 사용하여 gray컬러 출력하기
      2. 

```python
import cv2 
import matplotlib.pyplot as plt

# 1. gray 이미지로 바로 읽어오는방법
# 이미지 불러오기 
img = cv2.imread('./data/다운로드.jpeg', cv2.IMREAD_GRAYSCALE) # 이미지를 읽어올때 흑백으로 읽어오도록 설정 
plt.imshow(img, cmap = 'gray')
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
#2. 컬러이미지로 읽어서 gray 이미지로 변경해주는 방법 

img_color = cv2.imread('./data/다운로드.jpeg', cv2.IMREAD_COLOR)

# 색상변경 옵션 작성 
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)
plt.imshow(img_color)
plt.xticks([])
plt.yticks([])

plt.show()
```

### 이진이미지 만들기
- 이진이미지 : 0 과 255 로구성된이미지
- 주로 관심,비관심 영역 구분하고 싶을때 사용한다.

- - flag : 문턱값 적용 방법 (다양한 형태의 이진화 방법) 실제로는 위의 두값을 사용
  - cv2.THRESH_BINARY : 픽셀값이 문턱값보다 크면 value, 아니면 0을 할당
  - cv2.THRESH_BINARY_INV : 픽셀값이 문턱값보다 크면 0, 아니면 value를 할당 (inverse)
  - cv2.THRESH_TRUNC : 픽셀값이 문턱값보다 크면 문턱값, 아니면 픽셀값을 그대로 할당
  - cv2.THRESH_TOZERO : 픽셀값이 문턱값보다 크면 픽셀값, 아니면 0을 할당
  - cv2.THRESH_TOZERO_INV : 픽셀값이 문턱값보다 크면 0, 아니면 픽셀값을 할당

```python
# 흑백이미지로 불러오기 
gray = cv2.imread('./data/다운로드.jpeg', cv2.IMREAD_GRAYSCALE)

# 이진이미지로 변경 
# 문턱값(임계치) 기준을 두어 가진보다 작으면 0, 기준보다 크면 255로 만들어버린다.
# cv2.threshold( 이미지, 문턱값, 문턱값보다 클때 적용하는 값, 적용방법 ) 
_,thr1 = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)

# 첫번째 추출값 : 문턱값 
# 두번째 추출값 : 이진이미지 

# inverse(반전)  는 주로 ocr  글씨를 출력할때 사용한다
# 우리가 필요한건 글씨이기때문에 글씨를 검정색으로 출력해주기위함!
plt.subplot(2, 2, 1)  
_, thr1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thr1, cmap='gray')
plt.title('THRESH_BINARY_INV')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 2)   
_, thr1 = cv2.threshold(gray, 127, 255, cv2.THRESH_TRUNC)
plt.imshow(thr1, cmap='gray')
plt.title('THRESH_TRUNC')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 3)   
_, thr1 = cv2.threshold(gray, 127, 255, cv2.THRESH_TOZERO)
plt.imshow(thr1, cmap='gray')
plt.title('THRESH_TOZERO')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 2, 4)   
_, thr1 = cv2.threshold(gray, 127, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thr1, cmap='gray')
plt.title('THRESH_TOZERO_INV')
plt.xticks([]), plt.yticks([])
```
